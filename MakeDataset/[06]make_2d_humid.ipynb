```python
import warnings
from PIL import Image
from utils import *
warnings.filterwarnings('ignore')
```

```python
humider = {'강의지원': 50, '공용공간': np.nan, '교수연구': 60, '연구공간': 50, '학생자율': 60, '학술지원': 50, '행정지원': 60}

df = pd.read_csv('../temp/dataset_02[시간별목적별전력온습도].csv')
# df.fillna(0, inplace=True)
cols = list(df.columns[7:-1])

humid = df.drop(['use', 'temp', 'sum'], axis=1)
print('변경 전')
display(humid)

for key in humider.keys():
    humid[key] = humider[key]
    humid[key] -= humid['humid']

humid.drop('humid', axis=1, inplace=True)
print('변경 후')
humid.fillna(0, inplace=True)
humid = np.absolute(humid)
display(humid)
```

```python
building = pd.read_csv('../temp/dataset_03[1층].csv')
building.replace('0', 0, inplace=True)
building
```

```python
img_path = '../dataset/2d/img/humid'
npy_path = '../dataset/2d/npy/humid'

check_path(img_path)
check_path(npy_path)
```

```python
start = time.time()
# npy save
res = []
for idx in tqdm(humid.index):
    humid_b = building.copy()
    humid_b.replace(cols, humid.iloc[idx][4:], inplace=True)
    res.append(humid_b)
np.save(npy_path + '/2d_humid.npy', np.array(res))
end = time.time()
print('='*10, 'npy', '='*10)
print(f'저장 소요 시간 : {end-start:.2f}')


start = time.time()
max_val = np.max(humid[cols].max())
# png save
for idx in tqdm(humid.index):
    humid_b = building.copy()
    humid_b.replace(cols, humid.iloc[idx][4:], inplace=True)
    img = Image.fromarray((humid_b.to_numpy()/max_val)*255).convert('L')
    img.save(img_path + f'/{int(humid.iloc[idx]["year"])}_{int(humid.iloc[idx]["month"])}_{int(humid.iloc[idx]["day"])}_{int(humid.iloc[idx]["hour"])}.png', 'png')
end = time.time()
print('='*10, 'png', '='*10)
print(f'저장 소요 시간 : {end-start:.2f}')
```
