# temp with 결측치 처리(0)

```python
temper = {'강의지원': 23, '공용공간': np.nan, '교수연구': 20, '연구공간': 14, '학생자율': 18, '학술지원': 16, '행정지원': 18}

df = pd.read_csv('../temp/dataset_02[시간별목적별전력온습도].csv')
# df.fillna(0, inplace=True)
cols = list(df.columns[7:-1])

temp = df.drop(['use', 'humid', 'sum'], axis=1)
print('변경 전')
display(temp)

for key in temper.keys():
    temp[key] = temper[key]
    temp[key] -= temp['temp']

temp.drop('temp', axis=1, inplace=True)
print('변경 후')
temp.fillna(0, inplace=True)
temp = np.absolute(temp)
display(temp)
```

```python
building = pd.read_csv('../temp/dataset_03[1층].csv')
building.replace('0', 0, inplace=True)
building
```

```python
img_path = '../dataset/2d/img/temp'
npy_path = '../dataset/2d/npy/temp'

check_path(img_path)
check_path(npy_path)
```

```python
start = time.time()
# npy save
res = []
for idx in tqdm(temp.index):
    temp_b = building.copy()
    temp_b.replace(cols, temp.iloc[idx][4:], inplace=True)
    res.append(temp_b)
np.save(npy_path + '/2d_temp.npy', np.array(res))
end = time.time()
print('='*10, 'npy', '='*10)
print(f'저장 소요 시간 : {end-start:.2f}')


start = time.time()
max_val = np.max(temp[cols].max())
# png save
for idx in tqdm(temp.index):
    temp_b = building.copy()
    temp_b.replace(cols, temp.iloc[idx][4:], inplace=True)
    img = Image.fromarray((temp_b.to_numpy()/max_val)*255).convert('L')
    img.save(img_path + f'/{int(temp.iloc[idx]["year"])}_{int(temp.iloc[idx]["month"])}_{int(temp.iloc[idx]["day"])}_{int(temp.iloc[idx]["hour"])}.png', 'png')
end = time.time()
print('='*10, 'png', '='*10)
print(f'저장 소요 시간 : {end-start:.2f}')
```
